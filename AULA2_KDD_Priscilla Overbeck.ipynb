```python
import pandas as pd
import numpy as np
```

```python
df_estados = pd.read_csv('estados.csv', sep=',', index_col='id_estado')
print(df_estados.head(5))
```

```python
df_arenas = pd.read_csv('arenas.csv', sep=',', index_col='id_arena')
print(df_arenas.head(5))
```

```python
df_clubes = pd.read_csv('clubes.csv', sep=',', index_col='id_clube')
print(df_clubes.head(5))
```

```python
df_jogos2013 = pd.read_csv('jogos_2013.csv', sep=',')
df_jogos2013['ano'] = 2013
print(df_jogos2013.head(5))
```

```python
df_jogos2014 = pd.read_csv('jogos_2014.csv', sep=',')
df_jogos2014['ano'] = 2014
print(df_jogos2014.head(5))
```

```python
df_jogos2015 = pd.read_csv('jogos_2015.csv', sep=',')
df_jogos2015['ano'] = 2015
print(df_jogos2015.head(5))
```

```python
df_jogos2016 = pd.read_csv('jogos_2016.csv', sep=',')
df_jogos2016['ano'] = 2016
print(df_jogos2016.head(5))
```

```python
df_jogos2017 = pd.read_csv('jogos_2017.csv', sep=',')
df_jogos2017['ano'] = 2017
print(df_jogos2017.head(5))
```

# Concatenando os df de jogos

```python
jogos_total = pd.concat([df_jogos2013, df_jogos2014, df_jogos2015, df_jogos2016, df_jogos2017], axis = 0)
print(jogos_total.head(5))
```

```python
final = pd.merge(jogos_total, df_arenas, how='left', on=['id_arena'])
final = final.rename(columns={'desc_arena': 'arena'})
print(final.head(5))
```

```python
final = pd.merge(final, df_clubes, how='left', left_on=['id_clube1'], right_on = ['id_clube'])
final = final.rename(columns={'desc_clube': 'clube 1'})
print(final.head(5))
```

```python
final = pd.merge(final, df_clubes, how='left', left_on=['id_clube2'], right_on = ['id_clube'])
final = final.rename(columns={'desc_clube': 'clube 2'})
print(final.head(5))
```

```python
final = pd.merge(final, df_clubes, how='left', left_on=['id_clube_vencedor'], right_on = ['id_clube'])
final = final.rename(columns={'desc_clube': 'clube vencedor'})
print(final.head(5))
```

```python
final = pd.merge(final, df_estados, how='left', left_on=['id_estado_clube1'], right_on = ['id_estado'])
final = final.rename(columns={'desc_estado': 'estado clube 1'})
print(final.head(5))
```

```python
final = pd.merge(final, df_estados, how='left', left_on=['id_estado_clube2'], right_on = ['id_estado'])
final = final.rename(columns={'desc_estado': 'estado clube 2'})
print(final.head(5))
```

```python
final = pd.merge(final, df_estados, how='left', left_on=['id_estado_vencedor'], right_on = ['id_estado'])
final = final.rename(columns={'desc_estado': 'estado vencedor'})
print(final.head(5))
```

## a) Qual o clube que mais marcou gols em todas as temporadas?

```python
gols_clube1 = pd.DataFrame(final.groupby(['clube 1'])['qtd_gols_clube1'].sum())
gols_clube1.rename(columns={'qtd_gols_clube1':'qtd_gols'}, inplace=True)
gols_clube2 = pd.DataFrame(final.groupby(['clube 2'])['qtd_gols_clube2'].sum())
gols_clube2.rename(columns={'qtd_gols_clube2':'qtd_gols'}, inplace=True)
gols_total = pd.DataFrame(pd.concat([gols_clube1, gols_clube2], axis =0, sort=True))
gols_total = pd.DataFrame(gols_total.groupby(gols_total.index)['qtd_gols'].sum())
gols_total.sort_values('qtd_gols', ascending=False)[:1]
```

## b) Qual o clube que mais marcou gols na temporada de 2015

```python
final_2015 = final[final['ano'] == 2015]
gols_clube1_2015 = pd.DataFrame(final_2015.groupby(['clube 1','ano'])['qtd_gols_clube1'].sum())
gols_clube1_2015.rename(columns={'qtd_gols_clube1':'qtd_gols'}, inplace=True)
gols_clube2_2015 = pd.DataFrame(final_2015.groupby(['clube 2','ano'])['qtd_gols_clube2'].sum())
gols_clube2_2015.rename(columns={'qtd_gols_clube2':'qtd_gols'}, inplace=True)
gols_total_2015 = pd.DataFrame(pd.concat([gols_clube1_2015, gols_clube2_2015], axis =0, sort=True))
gols_total_2015 = pd.DataFrame(gols_total_2015.groupby(gols_total_2015.index)['qtd_gols'].sum())
gols_total_2015.sort_values('qtd_gols', ascending=False)[:1]
```

## c) Qual o jogo onde houveram mais gols, em todas as temporadas?

```python
mais_gols = pd.DataFrame(final, columns = ['total_gols','data', 'dia', 'horario', 'rodada',  'qtd_gols_clube1',
       'qtd_gols_clube2', 'arena', 'clube 1', 'clube 2', 'clube vencedor',
       'sigla_x', 'estado clube 1', 'sigla_y', 'estado clube 2', 'sigla',
       'estado vencedor', 'ano'])
mais_gols['total_gols'] = mais_gols['qtd_gols_clube1'] + mais_gols['qtd_gols_clube2']
mais_gols.sort_values('total_gols', ascending=False)[:1]
```

## d) Qual foi a maior goleada de todas as temporadas? Maior goleada nesse contexto é igual a maior diferença de gols entre o que um time marcou em relação ao outro.

```python
maior_goleada = pd.DataFrame(final, columns = ['diferença de gols','qtd_gols_clube1','qtd_gols_clube2','data',
        'dia', 'horario', 'rodada',  'arena', 'clube 1', 'clube 2', 'clube vencedor',
       'sigla_x', 'estado clube 1', 'sigla_y', 'estado clube 2', 'sigla',
       'estado vencedor', 'ano'])
maior_goleada['diferença de gols'] = maior_goleada['qtd_gols_clube1'] - maior_goleada['qtd_gols_clube2']
maior_goleada.sort_values('diferença de gols', ascending=False)
```

## e) Qual foi o estado mais vitorioso (quantidade de partidas) em 2016?

```python
estado_2016 = final[final['ano'] == 2016]
estado_2016 = pd.DataFrame(estado_2016.groupby(['estado vencedor'])['sigla'].count())
estado_2016 = estado_2016.rename(columns={'sigla': 'partidas vencidas'})
estado_2016.sort_values(['partidas vencidas'], ascending=False)[0:2]
```
